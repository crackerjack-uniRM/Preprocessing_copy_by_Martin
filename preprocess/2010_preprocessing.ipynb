### 3：剔除管理层平均工资比普通员工平均工资低的公司,因为这类公司披露的高管薪酬可能仅是津贴,或者数据错误：

```python
full_data.columns
```

```python
full_data.columns = ['证券代码', '证券简称', '所属申万行业名称一级行业', '董事会人数',
       '独立董事人数', '非独立董事人数',
       '公司监事', '高管人数',
       '支付给职工以及为职工支付的现金',
       '管理层年度薪酬总额', '资产总计',
       '员工总数', '净现金流', '资产负债率',
       '营业收入', '公司属性',
       '大股东持股比例', 'Stkcd', 'EndDate', 'Same',
       'Symbol', 'Enddate', 'StatisticalCaliber', 'UnpaidNumber',
       'check_ST', 'check_*ST']
full_data.columns
```

```python
full_data.shape
```

```python
# 管理层平均工资 = 管理层年度薪酬总数/（非独立董事人数+公司监事+高管人数-UnpaidNumber）
# 普通员工平均工资 = （支付给职工以及为职工支付的现金-管理层年度薪酬总数）/ （员工总数-（非独立董事+公司监事+高管人数））
```

```python
full_data['人数']=full_data['公司监事'].apply(lambda x: len(x.split(',')))
```

```python
full_data['new'] = full_data['支付给职工以及为职工支付的现金'] / full_data['人数']
```
